```julia
include("../src/HJB.jl")
```

```julia
# create the grid
x1s = range(0.0, 2.0, length=128)
x2s = range(-1.0, 2.0, length=256)

grid = HJB.Grid((x1s, x2s))
```

```julia
function plot_grid!(data, grid::HJB.Grid{2, F}) where {F}
    # for now only supports plotting 2 dimensional problems

    x1s = HJB.get_axes(grid, 1)
    x2s = HJB.get_axes(grid, 2)

    heatmap!(x1s, x2s, data')

end
```

```julia
# define the dynamics
function f(t, x, u)
    return @SVector [x[2], u[1]]
end
```

```julia
# define the Hamiltonian
function H(t, x, V, DxV, params)
    u1 = @SVector [1.0]
    u2 = @SVector [-1.0]

    H1 = DxV' * f(t, x, u1)
    H2 = DxV' * f(t, x, u2)

    return min(0, max(H1, H2))
end
```

```julia
# define the constraint function
function l(x)
    return 1.5 - x[1]
    
    # maxX = 1.5 - x[1]
    # maxV = 1.5 - x[2]
    # minV = x[2] + 0.5
    # return min(maxX, maxV, minV)
    
end
```

```julia
# construct the HJB problem
tspan = (0.0, 2.0)
prob = HJB.HJBProblem(H, l, tspan)
```

```julia
VF = HJB.allocate_grid(Float64, grid, l)
dV = similar(VF);
```

```julia
RHS! = HJB.get_ODE_RHS(prob, grid, HJB.Direct(),HJB.Simple())
```

```julia
# solve the ODE
@time sol = HJB.solve_MOL(prob, grid; gradMethod=HJB.WENO(), timeMethod=HJB.Direct(), reltol=1e-1, abstol=1e-1)
```

```julia
@time sol = HJB.solve_MOL(prob, grid; gradMethod=HJB.Simple(), timeMethod=HJB.Direct(), reltol=1e-8, abstol=1e-8)
```

```julia
sol.sol
```

```julia
ind = 1 # length(sol.sol.t)

p1 = plot()
plot_grid!(sol.sol.u[ind], grid)
title!("t = $(sol.sol.t[ind])")

p2 = plot()
plot_grid!(sol.sol.u[ind] .>= 0.0, grid)
title!("t = $(sol.sol.t[ind])")

plot!(p1, p2)
```
